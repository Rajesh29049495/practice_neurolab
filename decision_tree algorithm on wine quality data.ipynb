#### **NOTE: for the decision tree scaling is not required**

```python
X = df.drop(['quality'],axis= 1)
```

```python
y = df['quality']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)   ###this gives model accuracy w.r.t. to train data
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_predict)   ##checking model's accuracy w.r.t. test data
```

**this model is a overfitted model**{just noting this thing for now, because right now i am just seeing how to form the decision tree model,,,later on will see how to resolve this and improve the accuracy using hyperparameter tuning }

###### Hyper parameter tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_param ={
    'criterion':['gini','entropy'],
    'max_depth':range(2,32,1),
    'min_samples_leaf':range(1,10,1),
    'min_samples_split':range(2,10,1),
    'splitter':['best','random']
}
```

```python
grid_search= GridSearchCV(estimator=model,param_grid=grid_param, cv=5)   ##if i would have included the 'verbose' parameter inside thi sthen whhile training th emodel it would have shown the no. of fits it is running with the mentioned hyper parameter inside 'param_gris'
```

```python
grid_search.fit(X_train,y_train)
```

```python
##it takes alot of time, sometime even 5-6 hours if data is too big
```

```python
grid_search.best_params_   ###gives the best combination of hyperparameters
```

```python
## now we going to form the model with that best parameter

model_with_best_para= DecisionTreeClassifier(criterion= 'gini',
 max_depth = 5,
 min_samples_leaf= 4,
 min_samples_split= 4,
 splitter= 'random') 
```

```python
## train the model

model_with_best_para.fit(X_train,y_train)
```

```python
model_with_best_para.score(X_train,y_train)
```

```python
y_pred2=model_with_best_para.predict(X_test)
```

```python
accuracy_score(y_test,y_pred2)
```

**accuracy is low i.e., decision tree is not good for this, i think random forest will be the best for this**

###### Hyperparameter tuning

```python
### time complexity in this case is high, as by default only 100 of dts are being made,,, then if do the hyperparameter tuning,then the time complexity will boviously going to increase.
```

```python
grid_param = {
    "n_estimators" : [90,100,115,130],
    'criterion': ['gini', 'entropy'],
    'max_depth' : range(2,20,1),
    'min_samples_leaf' : range(1,10,1),
    'min_samples_split': range(2,10,1),
    'max_features' : ['auto','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
grid_searh=GridSearchCV(estimator=Rf_model,param_grid=grid_param,cv=3,verbose=2,n_jobs=-1)
```

```python
grid_searh.fit(X_train,y_train)
```

```python
#verbose give validation accuracy w.r.t each iteration,,it gives configuration detail of the algorithm
#njobs will tell how much cpu being consumed
```

```python
grid_searh.best_params_
```

```python
Rf_model_with_best_params=RandomForestClassifier(criterion='gini',max_depth= 5,max_features= 'auto',min_samples_leaf= 1,min_samples_split= 5,n_estimators=115)
```

```python
Rf_model_with_best_params.fit(X_train,y_train)
```

## bagging algorithm
{uses bagging concept, we bag homogenous algorithms and form model out of them, like the decission trees are bagged in random forest algorithm}

```python
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
from sklearn.datasets import make_classification

model_bagging_svc = BaggingClassifier(base_estimator=SVC(),n_estimators=50, random_state=0).fit(X_train, y_train)
```

```python
y_predict_bagging= model_bagging_svc.predict(X_test)
```

```python
accuracy_score(y_test,y_predict_bagging)
```
